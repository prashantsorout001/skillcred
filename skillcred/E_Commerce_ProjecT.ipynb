## Q 8. Create a new dataframe to include the Brand specific information as stated:
## i.	total number of orders placed per Brand
## ii.	total retail_price per Brand
## iii.	total discounted_price per Brand, and
## iv.	total BrandRevenue generated per Brand.

## Also, draw a pairplot using these four features. (6 marks)


### Ans 8

```python
def calculate_company_revenue(discounted_price):
    if (discounted_price > 600):
        return discounted_price * 0.25
    elif (discounted_price > 350):
        return discounted_price * 0.15
    elif (discounted_price > 100):
        return discounted_price * 0.10
    else:
        return discounted_price * 0.05

# Calculate company revenue for each order
ECom_Data['company_revenue'] = ECom_Data['discounted_price'].apply(calculate_company_revenue)

# Calculate brand revenue for each order
ECom_Data['brand_revenue'] = ECom_Data['discounted_price'] - ECom_Data['company_revenue']

# Calculate total number of orders per Brand
brand_total_orders = ECom_Data.groupby('Brand')['Customer_uniq_id'].size().reset_index(name='total_orders')

# Calculate total retail price per Brand
brand_total_retail_price = ECom_Data.groupby('Brand')['retail_price'].sum().reset_index(name='total_retail_price')

# Calculate total discounted price per Brand
brand_total_discounted_price = ECom_Data.groupby('Brand')['discounted_price'].sum().reset_index(name='total_discounted_price')

# Calculate total brand revenue per Brand
brand_total_revenue = ECom_Data.groupby('Brand')['brand_revenue'].sum().reset_index(name='total_brand_revenue')


Brand_Info = pd.merge(brand_total_orders, brand_total_retail_price, on='Brand')
Brand_Info = pd.merge(Brand_Info, brand_total_discounted_price, on='Brand')
Brand_Info = pd.merge(Brand_Info, brand_total_revenue, on='Brand')

print("Brand Specific Information:")
print(Brand_Info)

sns.pairplot(Brand_Info.drop('total_orders', axis=1))
plt.suptitle('Pairplot of Brand Specific Information', y=1.02)
plt.show()
```

## Q 9.Compare performance regionwise

## a)	Draw a lineplot for the monthly Revenue of E-Commerce Company for each region separately.  (4 marks)

## b)	Identify the best and the worst performing months for each region. (2 marks)

#### Note: Only those days with actual orders(Order_Date) placed are present in the dataset. Assuming there were no orders on other days. Also, show Month and Year on x-axis.

### Ans 9 a)

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
ECom_Data = pd.read_csv("E-Commerce.csv")

# Function to calculate company revenue based on discounted price
def calculate_company_revenue(discounted_price):
    if discounted_price > 600:
        return discounted_price * 0.25
    elif discounted_price > 350:
        return discounted_price * 0.15
    elif discounted_price > 100:
        return discounted_price * 0.10
    else:
        return discounted_price * 0.05

# Calculate company revenue for each order
ECom_Data['company_revenue'] = ECom_Data['discounted_price'].apply(calculate_company_revenue)

# Calculate brand revenue for each order
ECom_Data['brand_revenue'] = ECom_Data['discounted_price'] - ECom_Data['company_revenue']

# Convert Order_Date to datetime format
ECom_Data['Order_Date'] = pd.to_datetime(ECom_Data['Order_Date'], format='%d-%m-%Y')

# Extract Month and Year from Order_Date
ECom_Data['Month'] = ECom_Data['Order_Date'].dt.month
ECom_Data['Year'] = ECom_Data['Order_Date'].dt.year

# Group by Region, Year, and Month and calculate total brand revenue
grouped_revenue = ECom_Data.groupby(['Region', 'Year', 'Month']).agg({'brand_revenue': 'sum'}).reset_index()

# Draw lineplot for monthly revenue of each region separately
plt.figure(figsize=(12, 6))

regions = grouped_revenue['Region'].unique()

for region in regions:
    region_data = grouped_revenue[grouped_revenue['Region'] == region]
    plt.plot(region_data['Month'], region_data['brand_revenue'], marker='o', label=region)

plt.title('Monthly Revenue of E-Commerce Company by Region')
plt.xlabel('Month')
plt.ylabel('Monthly Revenue')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.legend(title='Region')
plt.tight_layout()
plt.show()
```

### Ans 9 b)

```python
for region in regions:
    region_data = grouped_revenue[grouped_revenue['Region'] == region]
    best_month = region_data.loc[region_data['brand_revenue'].idxmax()]
    worst_month = region_data.loc[region_data['brand_revenue'].idxmin()]
    print(f"For {region}:")
    print(f"Best Performing Month: {best_month['Month']}-{best_month['Year']} with revenue ${best_month['brand_revenue']:.2f}")
    print(f"Worst Performing Month: {worst_month['Month']}-{worst_month['Year']} with revenue ${worst_month['brand_revenue']:.2f}")
    print()
```
